**Exercício 1:** construa uma lista contendo 100 números, em ordem crescente.

```python
l = []                       # 1
for i in range(100):
    l += [i]

l = [i for i in range(100)]  # 2

l = np.arange(100)           # 3

print(l)
```

**Exercício 2:** some todos os valores contidos na lista anterior.

```python
sum(l)
```

**Exercício 3:** re-formate a lista criada anteriormente em uma matriz de tamanho $10\times 10$.

```python
m = []                      # 1
for i in range(0, 100, 10):
    m += [l[i:i+10]]
m = np.asarray(m)

m = l.reshape(10, 10)       # 2

print(m)
```

**Exercício 4:** múltiplique a matriz anterior por sua transposta.

```python
m.dot(m.T).tolist()
```

**Exercício 5:** construa uma matriz de tamanho $3\times 3$, contendo $1s$ sobre a diagonal principal e $0$ nas demais posições (i.e., a matriz `eye`).

```python
e = np.zeros((3, 3))            # 1
e[0, 0] = e[1, 1] = e[2, 2] = 1

e = np.eye(3, 3)                # 2

print(e)
```

**Exercício 6:** múltiplique a matriz anterior por sua transposta.

```python
e.dot(e.T)
```

**Exercício 7:** randomize a posição dos elementos da lista construída no item 1.

```python
np.random.shuffle(l)

print(l)
```

**Exercício 7:** identifique os índices dos três maiores elementos da lista no item 7.

```python
indices = np.argsort(l)[-3:]

print('indices:', indices,
      'valores:', l[indices])
```

**Exercício 8:** construa uma matriz $P$ de tamanho $78\times 10$, contendo valores aleatórios contídos no intervalo $[0, 1)$.

```python
P = np.random.rand(78, 10)

P[:10].round(1)
```

**Exercício 9:** construa um vetor $V = [v_0, v_1, ..., v_n]$ contendo exatamente $78$ posições, tal que $v_i$ representa o índice referente ao maior valor contido na linha $i$ da matriz $P$.

```python
v = []                        # 1
for i in range(71):
    v.append(np.argmax(P[i]))

v = [np.argmax(r) for r in P] # 2

v = np.argmax(P, axis=1)      # 3

print(v)
```

**Exercício 10:** calcule a média e o desvio padrão de cada coluna da matriz $P$.

```python
print('média:', P.mean(axis=0).round(2))
print('desvio:', P.std(axis=0).round(2))
```
